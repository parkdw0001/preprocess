```python
df3=pd.read_csv('./서울특별시 성북구_흡연단속 현황2 - 복사본.csv')
```

```python
df3.columns
```

```python
pd.pivot_table(df3, index="권역명",columns='연도',aggfunc='sum', fill_value=0)
```

```python
pivot_df=pd.pivot_table(df3, index="권역명",columns='연도',aggfunc='sum', fill_value=0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'

pivot_df.plot(kind='bar', figsize=(12,7))
plt.title('서울특별시 성북구 흡연단속 현황')
plt.ylabel('적발건수')
plt.xlabel('권역명')
plt.show()
```
